```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Veriyi oku
file_path = "screentimeexcel.xlsx"
df = pd.read_excel(file_path)

# 'None' deerini 'No Event' olarak deitir ve eksik verileri 'No Event' ile doldur
df['Academic_Events'] = df['Academic_Events'].replace("None", "No Event").fillna("No Event")

# LinkedIn ve TikTok ekran sresi stunlarn doru ekilde saysal verilere dntr
df['Linkedin_ScreenTime (minutes)'] = pd.to_numeric(df['Linkedin_ScreenTime (minutes)'], errors='coerce')
df['Total_ScreenTime TikTok (minutes)'] = pd.to_numeric(df['Total_ScreenTime TikTok (minutes)'], errors='coerce')

# NaN deerlerini 0 ile doldur (Varsa)
df['Linkedin_ScreenTime (minutes)'] = df['Linkedin_ScreenTime (minutes)'].fillna(0)
df['Total_ScreenTime TikTok (minutes)'] = df['Total_ScreenTime TikTok (minutes)'].fillna(0)

# TikTok iin ortalama ekran sresi pivot tablosu olutur
tiktok_pivot = df.pivot_table(
    index='Day_Type',  # Gn tipi (Weekday, Weekend)
    columns='Usage_Time TikTok',  # Kullanm zaman (Morning, Afternoon, Evening)
    values='Total_ScreenTime TikTok (minutes)',  # TikTok ekran sresi
    aggfunc='mean'  # Ortalama al
)

# LinkedIn iin ortalama ekran sresi pivot tablosu olutur
linkedin_pivot = df.pivot_table(
    index='Day_Type',  # Gn tipi (Weekday, Weekend)
    columns='Usage_Time_LinkedIn',  # Kullanm zaman (Morning, Afternoon, Evening)
    values='Linkedin_ScreenTime (minutes)',  # LinkedIn ekran sresi
    aggfunc='mean'  # Ortalama al
)

# TikTok pivot tablosunu grselletir
plt.figure(figsize=(8, 5))
sns.heatmap(tiktok_pivot, annot=True, cmap='Greens', fmt='.1f')
plt.title('Average TikTok Screen Time by Day Type and Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Day Type')
plt.tight_layout()
plt.show()

# LinkedIn pivot tablosunu grselletir
plt.figure(figsize=(8, 5))
sns.heatmap(linkedin_pivot, annot=True, cmap='Blues', fmt='.1f')
plt.title('Average LinkedIn Screen Time by Day Type and Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Day Type')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, ttest_ind

# Load the dataset
file_path = "screentimeexcel.xlsx"
df = pd.read_excel(file_path)

# Clean data
df['Academic_Events'] = df['Academic_Events'].replace("None", "No Event").fillna("No Event")
df['Linkedin_ScreenTime (minutes)'] = pd.to_numeric(df['Linkedin_ScreenTime (minutes)'], errors='coerce').fillna(0)
df['Total_ScreenTime TikTok (minutes)'] = pd.to_numeric(df['Total_ScreenTime TikTok (minutes)'], errors='coerce').fillna(0)
df['Usage_Time TikTok'] = df['Usage_Time TikTok'].str.strip()
df['Usage_Time_LinkedIn'] = df['Usage_Time_LinkedIn'].str.strip()
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['Day_of_Week'] = df['Date'].dt.dayofweek
df['Day_Type'] = df['Day_of_Week'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')

# Assign quarters
df['Quarter'] = df['Month'].apply(lambda m: f"Q{((m - 1) // 3) + 1}")

# List of comparisons (including updated H4 and new H9)
comparisons = [
    (
        "LinkedIn screen time: working vs non-working hours",
        "H: LinkedIn screen time is equal during working and non-working hours.\n"
        "H: LinkedIn screen time is significantly higher during working hours.",
        df[df['Usage_Time_LinkedIn'].isin(['Morning', 'Afternoon'])]['Linkedin_ScreenTime (minutes)'],
        df[~df['Usage_Time_LinkedIn'].isin(['Morning', 'Afternoon'])]['Linkedin_ScreenTime (minutes)']
    ),
    (
        "TikTok screen time: night on weekends vs other times",
        "H: TikTok usage is equal or lower at night on weekends compared to other times.\n"
        "H: TikTok usage is significantly higher at night on weekends.",
        df[(df['Usage_Time TikTok'] == 'Night') & (df['Day_Type'] == 'Weekend')]['Total_ScreenTime TikTok (minutes)'],
        df[~((df['Usage_Time TikTok'] == 'Night') & (df['Day_Type'] == 'Weekend'))]['Total_ScreenTime TikTok (minutes)']
    ),
    (
        "LinkedIn screen time: early vs late academic year",
        "H: LinkedIn screen time is equal or lower in the second half of the academic year.\n"
        "H: LinkedIn screen time is significantly higher in the second half.",
        df[df['Date'].dt.month < 7]['Linkedin_ScreenTime (minutes)'],
        df[df['Date'].dt.month >= 7]['Linkedin_ScreenTime (minutes)']
    ),
    (
        "TikTok screen time: Q4 vs Q1+Q2+Q3",
        "H: There is no significant difference in TikTok usage in Q4 compared to other periods.\n"
        "H: TikTok usage is significantly different in Q4.",
        df[df['Quarter'] == 'Q4']['Total_ScreenTime TikTok (minutes)'],
        df[df['Quarter'].isin(['Q1', 'Q2', 'Q3'])]['Total_ScreenTime TikTok (minutes)']
    ),
    (
        "LinkedIn screen time: job application vs non-application days",
        "H: LinkedIn screen time is equal or lower on job application days.\n"
        "H: LinkedIn screen time is significantly higher on job application days.",
        df[df['Jobs_Applied'] > 0]['Linkedin_ScreenTime (minutes)'],
        df[df['Jobs_Applied'] == 0]['Linkedin_ScreenTime (minutes)']
    ),
    (
        "LinkedIn screen time: exam vs non-exam days",
        "H: LinkedIn screen time is equal or higher during exam periods.\n"
        "H: LinkedIn screen time is significantly lower during exam periods.",
        df[~df['Academic_Events'].str.lower().str.contains('exam')]['Linkedin_ScreenTime (minutes)'],
        df[df['Academic_Events'].str.lower().str.contains('exam')]['Linkedin_ScreenTime (minutes)']
    ),
    (
        "TikTok screen time: exam vs non-exam days",
        "H: TikTok screen time is equal or higher during exam periods.\n"
        "H: TikTok screen time is significantly lower during exam periods.",
        df[~df['Academic_Events'].str.lower().str.contains('exam')]['Total_ScreenTime TikTok (minutes)'],
        df[df['Academic_Events'].str.lower().str.contains('exam')]['Total_ScreenTime TikTok (minutes)']
    ),
    (
        "TikTok screen time: Q4 vs Q1+Q2+Q3 (recheck)",
        "H: There is no significant difference in TikTok usage in Q4 compared to other periods.\n"
        "H: TikTok usage is significantly different in Q4.",
        df[df['Quarter'] == 'Q4']['Total_ScreenTime TikTok (minutes)'],
        df[df['Quarter'].isin(['Q1', 'Q2', 'Q3'])]['Total_ScreenTime TikTok (minutes)']
    ),
    (
        "LinkedIn screen time: Q4 vs Q1+Q2+Q3",
        "H: There is no significant difference in LinkedIn usage in Q4 compared to other periods.\n"
        "H: LinkedIn usage is significantly different in Q4.",
        df[df['Quarter'] == 'Q4']['Linkedin_ScreenTime (minutes)'],
        df[df['Quarter'].isin(['Q1', 'Q2', 'Q3'])]['Linkedin_ScreenTime (minutes)']
    )
]

# Correlation
tt_all = df["Total_ScreenTime TikTok (minutes)"].dropna()
li_all = df["Linkedin_ScreenTime (minutes)"].dropna()
corr_coef, p_corr = pearsonr(tt_all, li_all)

# Run hypothesis tests
results = []
for label, hypothesis, g1, g2 in comparisons:
    _, p = ttest_ind(g1, g2, nan_policy='omit')
    results.append((label, hypothesis, g1.mean(), g2.mean(), p))

results.append((
    "Correlation: TikTok vs LinkedIn screen time",
    "H: There is no correlation between TikTok and LinkedIn usage.\n"
    "H: There is a significant negative correlation between TikTok and LinkedIn usage.",
    tt_all.mean(), li_all.mean(), p_corr
))

# Interpret results
def interpret(label, explanation, m1, m2, p, is_corr=False):
    if is_corr:
        corr = tt_all.corr(li_all)
        if p < 0.05 and corr < 0:
            return f"{label}\n{explanation}\nResult: Reject H  Significant negative correlation (r = {corr:.2f})"
        elif p < 0.05:
            return f"{label}\n{explanation}\nResult: Reject H  Significant positive correlation (r = {corr:.2f})"
        else:
            return f"{label}\n{explanation}\nResult: Fail to reject H  No significant correlation (r = {corr:.2f})"
    else:
        if p < 0.05:
            if m1 > m2:
                return f"{label}\n{explanation}\nResult: Reject H  Increase observed ({m1:.1f} > {m2:.1f})"
            else:
                return f"{label}\n{explanation}\nResult: Reject H  Decrease observed ({m1:.1f} < {m2:.1f})"
        else:
            return f"{label}\n{explanation}\nResult: Fail to reject H  No significant difference ({m1:.1f}  {m2:.1f})"

# Plot results
plt.figure(figsize=(13, 13))
plt.text(0.5, 0.96, "Hypothesis Testing Results", ha='center', fontsize=16, weight='bold')

y = 0.92
for i, (label, explanation, m1, m2, p) in enumerate(results):
    is_corr = "Correlation" in label
    text = interpret(label, explanation, m1, m2, p, is_corr)
    plt.text(0.5, y, text, ha='center', fontsize=9, wrap=True)
    y -= 0.095

sig_count = sum(p < 0.05 for _, _, _, _, p in results)
summary = "Conclusion: Multiple significant behavioral patterns were found." if sig_count >= 4 else "Conclusion: Few patterns were statistically significant."
plt.text(0.5, 0.04, summary, ha='center', fontsize=11, color='darkred')

plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
#pearson correlation
import pandas as pd
from scipy.stats import pearsonr


file_path = "screentimeexcel.xlsx"
df = pd.read_excel(file_path)


df['Academic_Events'] = df['Academic_Events'].replace("None", "No Event").fillna("No Event")
df['Date'] = pd.to_datetime(df['Date'])


df['Linkedin_ScreenTime (minutes)'] = pd.to_numeric(df['Linkedin_ScreenTime (minutes)'], errors='coerce')
df['Total_ScreenTime TikTok (minutes)'] = pd.to_numeric(df['Total_ScreenTime TikTok (minutes)'], errors='coerce')
df['Jobs_Applied'] = pd.to_numeric(df['Jobs_Applied'], errors='coerce').fillna(0)
df['Day_of_Year'] = df['Date'].dt.dayofyear

# 1. Linkedin vs TikTok
r1, p1 = pearsonr(df['Linkedin_ScreenTime (minutes)'], df['Total_ScreenTime TikTok (minutes)'])

# 2. TikTok vs Exam Days (binary: Exam=1, No Event=0)
df['Exam_Day'] = df['Academic_Events'].apply(lambda x: 1 if 'exam' in x.lower() else 0)
r2, p2 = pearsonr(df['Exam_Day'], df['Total_ScreenTime TikTok (minutes)'])

# 3. Linkedin vs Exam Days
r3, p3 = pearsonr(df['Exam_Day'], df['Linkedin_ScreenTime (minutes)'])

# 4. Linkedin vs Job Applications
r4, p4 = pearsonr(df['Jobs_Applied'], df['Linkedin_ScreenTime (minutes)'])

# 5. Linkedin vs Time (Graduation Trend)
r5, p5 = pearsonr(df['Day_of_Year'], df['Linkedin_ScreenTime (minutes)'])

# 6. TikTok vs Time (Graduation Trend)
r6, p6 = pearsonr(df['Day_of_Year'], df['Total_ScreenTime TikTok (minutes)'])

# Sonular yazdr
print("1. LinkedIn vs TikTok:", f"r = {r1:.2f}, p = {p1:.4f}")
print("2. TikTok vs Exam Days:", f"r = {r2:.2f}, p = {p2:.4f}")
print("3. LinkedIn vs Exam Days:", f"r = {r3:.2f}, p = {p3:.4f}")
print("4. LinkedIn vs Job Applications:", f"r = {r4:.2f}, p = {p4:.4f}")
print("5. LinkedIn vs Time (Graduation):", f"r = {r5:.2f}, p = {p5:.4f}")
print("6. TikTok vs Time (Graduation):", f"r = {r6:.2f}, p = {p6:.4f}")

#r positive: positive correlation
#r negative: negative correlation
#p < 0.05 accept h1
```
